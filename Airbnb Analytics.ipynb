```python
cols = ['id','listing_url','name','neighbourhood','city','state','zipcode','latitude',
 'longitude','is_location_exact','bathrooms','bedrooms','price','weekly_price',
 'monthly_price','cleaning_fee','availability_30','availability_60',
 'availability_90','availability_365','review_scores_rating',
 'review_scores_accuracy','review_scores_cleanliness',
 'review_scores_checkin','review_scores_communication',
 'review_scores_location','review_scores_value']


Chicago = pd.read_csv('AirbnbChicago.csv',sep=',',error_bad_lines=False,usecols=cols)

LA = pd.read_csv('AirbnbLA.csv',sep=',',error_bad_lines=False,usecols=cols)

NYC = pd.read_csv('AirbnbNYC.csv',sep=',',error_bad_lines=False,usecols=cols)

SF = pd.read_csv('AirbnbSF.csv',sep=',',error_bad_lines=False,usecols=cols)

SD = pd.read_csv('AirbnbSanDiego.csv',sep=',',error_bad_lines=False,usecols=cols)

Oakland = pd.read_csv('AirbnbOakland.csv',sep=',',error_bad_lines=False,usecols=cols)
           
Chicago.head(2)
len(Chicago.columns)
```

```python
# Stack all datasets on top of each other
Chicago = pd.DataFrame(Chicago)
LA = pd.DataFrame(LA)
NYC = pd.DataFrame(NYC)
SF = pd.DataFrame(SF)
SD = pd.DataFrame(SD)
Oakland = pd.DataFrame(Oakland)

frames = (Chicago,LA,NYC,SF,SD,Oakland)
Result = pd.concat(frames)

Result['price'] = Result['price'].replace('[\$,)]','', regex=True)
Result['price'] = Result['price'].replace( '[(]','-',regex=True )
Result['price'] = Result['price'].astype(float)

Result['weekly_price'] = Result['weekly_price'].replace('[\$,)]','', regex=True)
Result['weekly_price'] = Result['weekly_price'].replace( '[(]','-',regex=True )
Result['weekly_price'] = Result['weekly_price'].astype(float)


Result['monthly_price'] = Result['monthly_price'].replace('[\$,)]','', regex=True)
Result['monthly_price'] = Result['monthly_price'].replace( '[(]','-',regex=True )
Result['monthly_price'] = Result['monthly_price'].astype(float)


Result['cleaning_fee'] = Result['cleaning_fee'].replace('[\$,)]','', regex=True)
Result['cleaning_fee'] = Result['cleaning_fee'].replace( '[(]','-',regex=True )
Result['cleaning_fee'] = Result['cleaning_fee'].astype(float)
```
